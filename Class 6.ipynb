```python
import numpy as np
import pandas as pd
df=np.random.randint(1,10,size=(1000,5))
df=pd.DataFrame(df,columns=["S1","S2","S3","S4",'S5']) # We can change the name of columns from index values to column names.
df
```

```python
df['Marks Obtained']= df[["S1","S2","S3","S4","S5"]].apply(lambda x:sum(x),axis=1) # Axis=1 is means that it will sum rows or without it would have sumed the columns vertically.
df # instead of manually adding each row using + sign just use lambda(one line function) to sum it all up.
```

```python
def my_addition(*x):
    return sum(x)
df["top marks"]=df[["S1","S2"]].apply(lambda x:my_addition(*x),axis=1)
df
```

```python
def result(*x):
    total=50
    obtained=sum(x)
    percentage=(obtained/total)*100
    Grade="I"
    if percentage>90:
        Grade="A*"
    elif percentage>=80:
        Grade="A"
    elif percentage>=70:
        Grade="B"
    elif percentage>=50:
        Grade="C"
    else:
        Grade="F"
    return Grade 

df['Grade']=df[["S1","S2","S3","S4","S5"]].apply(lambda x:result(*x),axis=1)
df
```

```python
# When we connect or open a file we have some access or modes of the file that we open in python. r mode means only read, r+ means reading and writing which means I can write in the file, w mode means opening a file for writing. a mode will mean append which will append the file with another. Read the modes from teh link above. With each mode it will also tell you the position of cursor which is important as you will know through it where writing in the file will happen in the start or in the end
```

```python
with open("abc.txt",'r+') as f: #r+ mode means read and write
    a=f.read()
    print(a)
```

```python
with open("abc.txt",'r+') as f: #r+ mode means read and write. If I don't write any mode then python will assume mode to be r which means just read and would not support anything else.
    a=list(reversed(f.readlines()))
    print(a,"A")
    b=list(reversed(["A","B","C"]))
    print(b,"B")
    c=a+b
    print(c,"C")
    d="".join(c)
    print(d,"D")

    f.seek(0)
    f.write(d) # you have now written over the file. You can see the file in the folder and you will see that file has what is written in d but initially you saved the file with the content in the cell above.
```

```python
#f.read() would open the file at once
#f.readline is generator function and would open the file using next keyword line by line
#.readlines would open a file in a list.
```

```python
with open("abc.txt","w") as f: #w mode means write only so it will not read the file and therefore you are getting not readable error.
    a=f.read()
    print(a) 
```
